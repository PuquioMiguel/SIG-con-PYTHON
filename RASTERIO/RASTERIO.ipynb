## GUARDAR

Guardar una banda

```python
# Banda 7: NEAR IR1
nir1 = image.read(7)
```

```python
NIR1_image = rasterio.open('DATASET/NIR.tif',
                            mode = 'w',
                            driver = 'Gtiff',
                            width = image.width,
                            height = image.height,
                            count = 1,
                            crs = image.crs,
                            transform = image.transform,
                            dtype = image.meta['dtype']               
                            )
NIR1_image.write(nir1, 1)
NIR1_image.close()
```

Cambiar el nombre de la banda al guardar

```python
red = image.read(5)
nombre_banda = "Red"
```

```python
red_image = rasterio.open('DATASET/Red.tif',
                          mode = 'w',
                          driver = 'Gtiff',
                          width = image.width,
                          height = image.height,
                          count = 1,
                          crs = image.crs,
                          transform = image.transform,
                          dtype = image.meta['dtype']               
                         )
red_image.write(red, 1)
red_image.set_band_description(1, nombre_banda)
red_image.close()
```

Guardar combinaciones de bandas

```python
img_rgb = image.read([5,3,2])
nombre_banda = ["Red", "Green", "Blue"]
```

```python
rgb = rasterio.open('DATASET/rgb.tif',
                          mode = 'w',
                          driver = 'Gtiff',
                          width = image.width,
                          height = image.height,
                          count = 3,
                          crs = image.crs,
                          transform = image.transform,
                          dtype = image.meta['dtype'], 
                          nodata=0
                         )
rgb.write(img_rgb)
rgb.set_band_description(1, nombre_banda[0])
rgb.set_band_description(2, nombre_banda[1])
rgb.set_band_description(3, nombre_banda[2])
rgb.close()
```

## INDICES DE VEGETACION

```python
# Covertir las bandas a numpy
nir = image.read(7).astype('float32')
red = image.read(5).astype('float32')
```

```python
# Calcular el indice NDVI
NDVI = (nir - red) / (nir + red)
```

```python
# Visualizar
rasterio.plot.show(NDVI, cmap="magma")
```

```python
# Guardar
ndviImage = rasterio.open('DATASET/ndvi.tif',
                          mode='w',
                          driver='Gtiff',
                          width=image.width,
                          height=image.height,
                          count=1,
                          crs=image.crs,
                          transform=image.transform,
                          dtype='float32',
                          nodata=255                
                         )
ndviImage.write(NDVI, 1)
ndviImage.close()
```
